# Aufgabe 2: Vergleich von Fahrradzählstellen


In diesem Notebook analysieren wir Fahrradzähldaten aus zwei verschiedenen Städten: Konstanz und Freiburg. Die Daten werden von den folgenden Quellen bezogen:

1. **Konstanz**: [Zählstelle Herose 2021](https://offenedaten-konstanz.de/sites/default/files/Zaehlstelle_Herose_2021_15min_3.csv) [Archive-Link](https://web.archive.org/web/20240714153903/https://offenedaten-konstanz.de/dataset/fahrrad-dauerz-hlstellen/resource/30d361a1-2a24-4a75-a98e-4ca3106a494e#{})
2. **Freiburg**: [Fahrradzählstellen Zeitreihe](https://fritz.freiburg.de:8443/duva2dcat/dataset/de-bw-freiburg-fahrrad-zaehlstellen_zeitreihe/content.csv) [Archive-Link](https://web.archive.org/web/20240714154049/https://www.daten-bw.de/web/guest/suchen/-/details/fahrrad-zahlstellen-zeitreihe)

Wir werden die Daten herunterladen, aufbereiten, visualisieren und vergleichen.

## Schritt 1: Einlesen der Daten
Im ersten Schritt laden wir die Datensätze herunter und lesen sie in Pandas DataFrames ein. Dies ermöglicht uns eine einfache und effiziente Datenmanipulation und Analyse.


```python
import pandas as pd

# URLs der Datensätze
url_konstanz = 'https://offenedaten-konstanz.de/sites/default/files/Zaehlstelle_Herose_2021_15min_3.csv'
url_freiburg = 'https://fritz.freiburg.de:8443/duva2dcat/dataset/de-bw-freiburg-fahrrad-zaehlstellen_zeitreihe/content.csv'

# Einlesen der Datensätze
data_konstanz = pd.read_csv(url_konstanz, delimiter=';')
data_freiburg = pd.read_csv(url_freiburg)

# Anzeigen der ersten fünf Zeilen
print("Konstanz Datensatz:\n", data_konstanz.head())
print("Freiburg Datensatz:\n", data_freiburg.head())
```

In diesem Schritt laden wir die Datensätze von den angegebenen URLs herunter und lesen sie in Pandas DataFrames ein. Dies ermöglicht uns eine einfache und effiziente Datenmanipulation und Analyse.

Wir verwenden die `pd.read_csv` Funktion von Pandas, um die CSV-Dateien einzulesen. Für den Datensatz aus Konstanz geben wir den `delimiter=';'` an, da die Werte durch Semikolons getrennt sind. Für den Freiburger Datensatz ist keine spezielle Angabe notwendig, da die Werte durch Kommata getrennt sind.

Nachdem die Daten erfolgreich eingelesen wurden, zeigen wir die ersten fünf Zeilen jedes Datensatzes an, um die Struktur und den Inhalt der Daten zu überprüfen.


## Schritt 2: Datenaufbereitung

In diesem Schritt bereiten wir die Daten auf, indem wir sie bereinigen und die notwendigen Transformationen durchführen. Dies umfasst die Konvertierung von Datums- und Zeitspalten in ein geeignetes Format und das Entfernen von fehlenden Werten.


```python
# Konstanz-Datensatz aufbereiten
data_konstanz['Zeit'] = pd.to_datetime(data_konstanz['Zeit'], format='%d.%m.%Y %H:%M')
data_konstanz['Fahrradbruecke'] = pd.to_numeric(data_konstanz['Fahrradbruecke'], errors='coerce')
data_konstanz['Fahrradbruecke stadteinwaerts'] = pd.to_numeric(data_konstanz['Fahrradbruecke stadteinwaerts'], errors='coerce')
data_konstanz['Fahrradbruecke stadtauswaerts'] = pd.to_numeric(data_konstanz['Fahrradbruecke stadtauswaerts'], errors='coerce')
data_konstanz.dropna(inplace=True)

# Sicherstellen, dass die Datetime-Spalte im Freiburger Datensatz im datetime-Format vorliegt
data_freiburg['Datetime'] = pd.to_datetime(data_freiburg['Datetime'])
data_freiburg['Anzahl Fahrräder'] = pd.to_numeric(data_freiburg['Anzahl Fahrräder'], errors='coerce')
data_freiburg.dropna(inplace=True)

# Anzeigen der aufbereiteten Daten
print("Aufbereiteter Konstanz Datensatz:\n", data_konstanz.head())
print("Aufbereiteter Freiburg Datensatz:\n", data_freiburg.head())
```

In diesem Schritt bereiten wir die Daten auf, indem wir die Datums- und Zeitspalten in ein geeignetes Format konvertieren und fehlende Werte entfernen. Dies ist notwendig, um die Daten für die nachfolgenden Analysen und Visualisierungen vorzubereiten.

### Konstanz-Datensatz

1. **Konvertierung der "Zeit"-Spalte**: Wir konvertieren die "Zeit"-Spalte in ein datetime-Format.
2. **Bereinigung der Daten**: Wir entfernen alle Zeilen mit fehlenden Werten.

### Freiburg-Datensatz

1. **Erstellen einer neuen "Datetime"-Spalte**: Wir kombinieren die "Datum" und "Stunde"-Spalten und konvertieren sie in ein datetime-Format.
2. **Bereinigung der Daten**: Wir entfernen alle Zeilen mit fehlenden Werten.

Nachdem die Daten erfolgreich aufbereitet wurden, zeigen wir die ersten fünf Zeilen jedes Datensatzes an, um die Aufbereitung zu überprüfen.

## Schritt 3: Visualisierung der Zeitreihen

In diesem Schritt erstellen wir einfache Plots für beide Zeitreihen. Dies ermöglicht uns, Trends und Muster in den Fahrradzähldaten für die beiden Städte Konstanz und Freiburg zu erkennen.


```python
import matplotlib.pyplot as plt

# Konstanz-Datensatz plotten
plt.figure(figsize=(12, 6))
plt.plot(data_konstanz['Zeit'], data_konstanz['Fahrradbruecke'], label='Fahrrad Count')
plt.xlabel('Zeit')
plt.ylabel('Anzahl Fahrräder')
plt.title('Fahrradzählungen in Konstanz 2021')
plt.legend()
plt.show()

# Freiburg-Datensatz plotten
plt.figure(figsize=(12, 6))
plt.plot(data_freiburg['Datetime'], data_freiburg['Anzahl Fahrräder'], label='Fahrrad Count')
plt.xlabel('Zeit')
plt.ylabel('Anzahl Fahrräder')
plt.title('Fahrradzählungen in Freiburg 2021')
plt.legend()
plt.show()
```

In diesem Schritt erstellen wir einfache Plots für beide Zeitreihen. Dies ermöglicht uns, Trends und Muster in den Fahrradzähldaten für die beiden Städte Konstanz und Freiburg zu erkennen.

### Konstanz-Datensatz

Wir plotten die Anzahl der Fahrräder, die an der Fahrradbrücke insgesamt, stadteinwärts und stadtauswärts gezählt wurden, gegen die Zeit.

### Freiburg-Datensatz

Wir plotten die Anzahl der Fahrräder, die an der Zählstelle in Freiburg gezählt wurden, gegen die Zeit.


## Schritt 4: Monatliche Aggregation und Vergleich

In diesem Schritt aggregieren wir die Daten monatsweise für die Monate des Jahres 2021 und stellen die beiden Städte Konstanz und Freiburg gegenüber. Dies ermöglicht uns, die Fahrradzähldaten der beiden Städte auf monatlicher Basis zu vergleichen.

```python
# Konstanz-Daten monatlich aggregieren
data_konstanz['Monat'] = data_konstanz['Zeit'].dt.to_period('M')
monthly_konstanz = data_konstanz.groupby('Monat')['Fahrradbruecke'].sum().reset_index()

# Freiburg-Daten monatlich aggregieren
data_freiburg['Monat'] = data_freiburg['Datetime'].dt.to_period('M')
monthly_freiburg = data_freiburg.groupby('Monat')['Anzahl Fahrräder'].sum().reset_index()

# Daten für das Jahr 2021 filtern
monthly_konstanz_2021 = monthly_konstanz[(monthly_konstanz['Monat'] >= '2021-01') & (monthly_konstanz['Monat'] <= '2021-12')]
monthly_freiburg_2021 = monthly_freiburg[(monthly_freiburg['Monat'] >= '2021-01') & (monthly_freiburg['Monat'] <= '2021-12')]

# Überprüfen der aggregierten Daten
print("Monatlich aggregierte Daten für Konstanz 2021:\n", monthly_konstanz_2021)
print("Monatlich aggregierte Daten für Freiburg 2021:\n", monthly_freiburg_2021)

# Vergleich der beiden Städte plotten
plt.figure(figsize=(12, 6))
plt.plot(monthly_konstanz_2021['Monat'].astype(str), monthly_konstanz_2021['Fahrradbruecke'], label='Konstanz Gesamt')
plt.plot(monthly_freiburg_2021['Monat'].astype(str), monthly_freiburg_2021['Anzahl Fahrräder'], label='Freiburg Gesamt')
plt.xlabel('Monat')
plt.ylabel('Anzahl Fahrräder')
plt.title('Monatliche Fahrradzählungen: Konstanz vs. Freiburg 2021')
plt.legend()
plt.show()
```

In diesem Schritt aggregieren wir die Daten monatsweise für die Monate des Jahres 2021 und stellen die beiden Städte Konstanz und Freiburg gegenüber. Dies ermöglicht uns, die Fahrradzähldaten der beiden Städte auf monatlicher Basis zu vergleichen.

### Konstanz-Datensatz

Wir aggregieren die Daten der Stadt Konstanz auf monatlicher Basis und filtern die Daten für das Jahr 2021. Dabei summieren wir die Anzahl der Fahrräder, die an der Fahrradbruecke insgesamt gezählt wurden.

### Freiburg-Datensatz

Wir aggregieren die Daten der Stadt Freiburg auf monatlicher Basis und filtern die Daten für das Jahr 2021. Dabei summieren wir die Anzahl der Fahrräder, die an der Zählstelle in Freiburg gezählt wurden.

### Vergleich der beiden Städte

Wir stellen die monatlich aggregierten Daten der beiden Städte gegenüber und plotten die Ergebnisse, um die Unterschiede und Ähnlichkeiten in den Fahrradzähldaten zu visualisieren.
